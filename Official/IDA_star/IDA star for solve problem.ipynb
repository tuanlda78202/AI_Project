# 5. Create 3 main warehouse

```python
def create_main_warehouse():
    random.seed(42)
    df=pd.read_excel('Data_Excel.xlsx')
    Latitude=list(df['Latitude'])
    Longitude=list(df['Longitude'])
    province=df['Province']
    position_main_warehouse1=[]
    position_main_warehouse2=[]
    position_main_warehouse3=[]
    for i in range(len(Latitude)):
        if Latitude[i] >20 :
            position_main_warehouse1.append(i)
        elif Latitude[i] >12:
            position_main_warehouse2.append(i)
        else:
            position_main_warehouse3.append(i)
    main_warehouse1=province[random.choice(position_main_warehouse1)].strip()
    main_warehouse2=province[random.choice(position_main_warehouse2)].strip()
    main_warehouse3=province[random.choice(position_main_warehouse3)].strip()
    list_main_warehouse=[main_warehouse1,main_warehouse2,main_warehouse3]
    return list_main_warehouse
print(create_main_warehouse())
```

# 6. IDA_star

```python
def IDA_star():
    global list_node_previous
    #define a threshold: theta =f(root_node) with f(n)=h(n)+g(n)
    threshold=list_distance[dict_index_province[start]][dict_index_province[goal]] 
    while True: # run infinity
        temp=search(start,0,threshold,parent='Null') #function search(node,g score,threshold)
        if temp=='FOUND': #if goal found
            return threshold
        threshold=temp  
def search(node,g,threshold,parent):  #recursive function
    f=g+list_distance[dict_index_province[node]][dict_index_province[goal]]
    if (f>threshold):#greater f encountered
        return f
    if (f<=threshold):
        list_node_previous[node]=parent
    if node==goal:   #Goal node found
        return 'FOUND'
    minn=10**10 #minn= Minimum integer
    for tempnode in nextnodes(node):
        #recursive call with next node as current node for depth search
        temp=search(tempnode,g+list_distance[dict_index_province[node]][dict_index_province[tempnode]],threshold,node)
        if temp=='FOUND':# if goal found
            return 'FOUND'
        if (temp<minn):# find the minimum of all f greater than threshold encountered  
            minn=temp
    return minn #//return the minimum f encountered greater than threshold
def nextnodes(node):
    return data_neighbour[node] #return list of all possible next nodes from node
def main_IDA_star():
    '''For easy handling, we divide it into 3 regions: 
the domain with longitude > 20,the domain with longitude from 12-20 and the region with longitude < 12
We discussed and found that such domain division reduces the running time and space complexity of the algorithm.
The two provinces that help us to effectively handle division are Thanh Hoa and Lam Dong.
'''
    global list_node_previous,start,goal
    list_node_previous={} # dictionary has key= expanded node, values= its parent node
    index_start=dict_index_province[start] # position of start in Province
    index_goal=dict_index_province[goal]   # position of goal in Province
    Lat_start=Latitude[index_start]        #Latitude of start
    Lat_goal=Latitude[index_goal]          # Latitude of goal
    list_solution=[]   
    if (Lat_start>12) and (Lat_goal>12):
        if (Lat_start<20) and (Lat_goal>20):#solve with latitude of start from 12 to 18 and latitude of goal > 20
            t=goal
            goal='Thanh Hoa'
            cost_path=IDA_star()                   
            check=goal
            list_solution.append(check)
            while check!=start:
                check=list_node_previous[check]
                list_solution.append(check)
            list_solution=list_solution[::-1] # save solution when we solved
            list_node_previous={}
            start='Thanh Hoa'
            goal=t
            cost_path+=IDA_star()
            check=goal
            list_solution1=[]
            list_solution1.append(check)
            while check!=start:
                check=list_node_previous[check]
                list_solution1.append(check)
            list_solution=list_solution[:-1]+list_solution1[::-1]
            list_node_previous={}
        else:
            cost_path=IDA_star()
            check=goal
            list_solution.append(check)
            while check!=start:
                check=list_node_previous[check]
                list_solution.append(check)
            list_solution=list_solution[::-1]
            list_node_previous={}
    if (Lat_start<12) and (Lat_goal<12):
        cost_path=IDA_star()
        check=goal
        list_solution.append(check)
        while check!=start:
            check=list_node_previous[check]
            list_solution.append(check)
        list_solution=list_solution[::-1]
        list_node_previous={}
    if (Lat_start>12) and (Lat_goal<12):
        t=goal
        goal='Lam Dong'
        cost_path=IDA_star()
        check=goal
        list_solution.append(check)
        while check!=start:
            check=list_node_previous[check]
            list_solution.append(check)
        list_solution=list_solution[::-1]
        list_node_previous={}
        start='Lam Dong'
        goal=t
        cost_path+=IDA_star()
        check=goal
        list_solution1=[]
        list_solution1.append(check)
        while check!=start:
            check=list_node_previous[check]
            list_solution1.append(check)
        list_solution=list_solution[:-1]+list_solution1[::-1]
        list_node_previous={}
    if (Lat_start<12) and (Lat_goal>12):
        t=goal
        goal='Lam Dong'
        cost_path=IDA_star()
        check=goal
        list_solution.append(check)
        while check!=start:
            check=list_node_previous[check]
            list_solution.append(check)
        list_solution=list_solution[::-1]
        list_node_previous={}
        if Lat_goal>20:
            start='Lam Dong'
            goal='Thanh Hoa'
            cost_path+=IDA_star()
            check=goal
            list_solution1=[]
            list_solution1.append(check)
            while check!=start:
                check=list_node_previous[check]
                list_solution1.append(check)
            list_solution=list_solution[:-1]+list_solution1[::-1]
            list_node_previous={}
            start='Thanh Hoa'
            goal=t
            cost_path+=IDA_star()
            check=goal
            list_solution1=[]
            list_solution1.append(check)
            while check!=start:
                check=list_node_previous[check]
                list_solution1.append(check)
            list_solution=list_solution[:-1]+list_solution1[::-1]
            list_node_previous={}
        else:    
            start='Lam Dong'
            goal=t
            cost_path+=IDA_star()
            check=goal
            list_solution1=[]
            list_solution1.append(check)
            while check!=start:
                check=list_node_previous[check]
                list_solution1.append(check)
            list_solution=list_solution[:-1]+list_solution1[::-1]
            list_node_previous={}
    return cost_path,list_solution # return cost distance and list_solution is the provinces it traverses from start to goal
'''Such domain division will help us to solve problems. 
The process function below will pass the input parameter to the problem'''
def process(start_province,goal_province):
    global start,goal
    start=start_province
    goal=goal_province
    distance_path,solution_path_list=main_IDA_star()
    return distance_path,solution_path_list
print(process('Ha Tinh','Vinh Phuc'))
```

```python
def main():
    list_main_warehouse=create_main_warehouse()
    print('Three of  main warehouse are ', list_main_warehouse)
    cost_truck = 35.000 
    cost_plane = 160.000 
    v_truck = 60 
    v_plane = 800 
    time_warehouse=2
    start_province=input('start province :')
    goal_province=input('end province : ')
    index_start=dict_index_province[start_province] # position of start in Province
    index_goal=dict_index_province[goal_province]   # position of goal in Province
    Lat_start=Latitude[index_start]        #Latitude of start
    Lat_goal=Latitude[index_goal]          # Latitude of goal
    list_cost=[]
    if (Lat_start>20) and (Lat_goal>20):
        a=dict_index_province[start_province]
        b=dict_index_province[goal_province]
        c=dict_index_province[list_main_warehouse[0]]
        if (list_distance[a][b] > list_distance[a][c]) and (list_distance[a][b] > list_distance[b][c]):
            cost=0
            list_cost=[]
            distance_path,solution_path_list=process(start_province,list_main_warehouse[0])
            cost=cost+distance_path/v_truck*cost_truck
            list_cost+=solution_path_list
            distance_path,solution_path_list=process(list_main_warehouse[0] ,goal_province)
            cost=cost+distance_path/v_truck*cost_truck
            list_cost=list_cost[:-1]+solution_path_list
            print('cost optimal solution for problem is ' ,cost)
            print('path optimal solution for problem is  ',' -> '.join(list_cost))
            return list_main_warehouse,list_cost # return for vizualization
        else:
            cost=0
            list_cost=[]
            distance_path,solution_path_list=process(start_province,goal_province)
            cost=cost+distance_path/v_truck*cost_truck
            list_cost+=solution_path_list
            print('cost optimal solution for problem is ' ,cost)
            print('path optimal solution for problem is  ',' -> '.join(list_cost))
            return  list_main_warehouse,list_cost # return for vizualization
        
    elif (12<Lat_start<20) and (12<Lat_goal<20):
        a=dict_index_province[start_province]
        b=dict_index_province[goal_province]
        c=dict_index_province[list_main_warehouse[1]]
        if (list_distance[a][b] > list_distance[a][c]) and (list_distance[a][b] > list_distance[b][c]):
            cost=0
            list_cost=[]
            distance_path,solution_path_list=process(start_province,list_main_warehouse[1])
            cost=cost+distance_path/v_truck*cost_truck
            list_cost+=solution_path_list
            distance_path,solution_path_list=process(list_main_warehouse[1] ,goal_province)
            cost=cost+distance_path/v_truck*cost_truck
            list_cost=list_cost[:-1]+solution_path_list
            print('cost optimal solution for problem is ' ,cost)
            print('path optimal solution for problem is  ',' -> '.join(list_cost))
            return list_main_warehouse,list_cost # return for vizualization
        else:
            cost=0
            list_cost=[]
            distance_path,solution_path_list=process(start_province,goal_province)
            cost=cost+distance_path/v_truck*cost_truck
            list_cost+=solution_path_list
            print('cost optimal solution for problem is ' ,cost)
            print('path optimal solution for problem is  ',' -> '.join(list_cost))
            return list_main_warehouse,list_cost # return for vizualization
        
    elif  (Lat_start<12) and (Lat_goal<12):
        a=dict_index_province[start_province]
        b=dict_index_province[goal_province]
        c=dict_index_province[list_main_warehouse[2]]
        if (list_distance[a][b] > list_distance[a][c]) and (list_distance[a][b] > list_distance[b][c]):
            cost=0
            list_cost=[]
            distance_path,solution_path_list=process(start_province,list_main_warehouse[2])
            cost=cost+distance_path/v_truck*cost_truck
            list_cost+=solution_path_list
            distance_path,solution_path_list=process(list_main_warehouse[2] ,goal_province)
            cost=cost+distance_path/v_truck*cost_truck
            list_cost=list_cost[:-1]+solution_path_list
            print('cost optimal solution for problem is ' ,cost)
            print('path optimal solution for problem is  ',' -> '.join(list_cost))
            return list_main_warehouse,list_cost # return for vizualization
        else:
            cost=0
            list_cost=[]
            distance_path,solution_path_list=process(start_province,goal_province)
            cost=cost+distance_path/v_truck*cost_truck
            list_cost+=solution_path_list
            print('cost optimal solution for problem is ' ,cost)
            print('path optimal solution for problem is  ',' -> '.join(list_cost))
            return list_main_warehouse,list_cost # return for vizualization
    else:
        if (Lat_start>20) and (20>Lat_goal>12):
            main_warehouse_pass1=list_main_warehouse[0]
            main_warehouse_pass2=list_main_warehouse[1]
            main_warehouse_pass3=list_main_warehouse[2]
        if (Lat_start>20) and (Lat_goal<12):
            main_warehouse_pass1=list_main_warehouse[0]
            main_warehouse_pass2=list_main_warehouse[2]
            main_warehouse_pass3=list_main_warehouse[1]
        if (12<Lat_start<20) and (Lat_goal>20):
            main_warehouse_pass1=list_main_warehouse[1]
            main_warehouse_pass2=list_main_warehouse[0]
            main_warehouse_pass3=list_main_warehouse[2]
        if (12<Lat_start<20) and (Lat_goal<12):
            main_warehouse_pass1=list_main_warehouse[1]
            main_warehouse_pass2=list_main_warehouse[2]
            main_warehouse_pass3=list_main_warehouse[0]
        if (Lat_start<12) and (Lat_goal>20):
            main_warehouse_pass1=list_main_warehouse[2]
            main_warehouse_pass2=list_main_warehouse[0]
            main_warehouse_pass3=list_main_warehouse[1]
        if (Lat_start<12) and (12<Lat_goal<20):
            main_warehouse_pass1=list_main_warehouse[2]
            main_warehouse_pass2=list_main_warehouse[1]
            main_warehouse_pass3=list_main_warehouse[0]
        list_cost=[]
        # go through 1 main warehouse
        cost=0
        list_solution_1_main_warehouse=[]
        distance_path,solution_path_list=process(start_province,main_warehouse_pass1)
        cost=cost+distance_path/v_truck*cost_truck
        list_solution_1_main_warehouse+=solution_path_list
        
        distance_path,solution_path_list=process(main_warehouse_pass1,goal_province)
        cost=cost+distance_path/v_truck*cost_truck
        list_solution_1_main_warehouse= list_solution_1_main_warehouse[:-1]+solution_path_list
        list_cost.append([cost,list_solution_1_main_warehouse])
        # go throught 2 main warehouse
        cost=0
        #start -> main warehouse 1,2
        list_solution_2_main_warehouse=[]
        distance_path,solution_path_list=process(start_province,main_warehouse_pass1)
        cost=cost+distance_path/v_truck*cost_truck
        list_solution_2_main_warehouse+=solution_path_list
        # main warehouse1 - main warehouse 2
        a=main_warehouse_pass1
        b=main_warehouse_pass2
        cost+=list_distance[dict_index_province[a]][dict_index_province[b]]/v_plane*cost_plane+time_warehouse*cost_plane
        # main warehouse 2 -> goal
        distance_path,solution_path_list=process(main_warehouse_pass2,goal_province)
        cost=cost+distance_path/v_truck*cost_truck
        list_solution_2_main_warehouse+=solution_path_list
        # append to list_cost 
        list_cost.append([cost,list_solution_2_main_warehouse])
        # go throught 2 main warehouse 1,3
        cost=0
        #start -> main warehouse 1
        list_solution_3_main_warehouse=[]
        distance_path,solution_path_list=process(start_province,main_warehouse_pass1)
        cost=cost+distance_path/v_truck*cost_truck
        list_solution_3_main_warehouse+=solution_path_list
        # main warehouse1 - main warehouse 3
        a=main_warehouse_pass1
        b=main_warehouse_pass3
        cost+=list_distance[dict_index_province[a]][dict_index_province[b]]/v_plane*cost_plane+time_warehouse*cost_plane
        # main warehouse 3 -> goal
        distance_path,solution_path_list=process(main_warehouse_pass3,goal_province)
        cost=cost+distance_path/v_truck*cost_truck
        list_solution_3_main_warehouse+=solution_path_list
        # append to list_cost 
        list_cost.append([cost,list_solution_3_main_warehouse])
        # print
        list_cost_sort=sorted(list_cost)
        optimal_solution=list_cost_sort[0]
        print('cost optimal solution for problem is ' ,optimal_solution[0])
        print('path optimal solution for problem is  ',' -> '.join(optimal_solution[1]))
        return list_main_warehouse,optimal_solution[1] # return for vizualization
list_main_warehouse,path_optimal_solution=main()  # use for vizualization
end_time=time.time()
print('Running time is ',end_time-start_time)
#print(path_optimal_solution)
```
